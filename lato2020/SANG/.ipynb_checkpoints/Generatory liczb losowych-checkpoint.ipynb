## 1. Generator von Neumanna

```python
import matplotlib.pyplot as plt
import numpy as np
import time
import random
from scipy import stats
import tqdm
```

```python
def vonNeumann(seed, N, iterations, generated_numbers = [], it = 0):
    '''
    seed - liczba początkowa, determinuje kolejne liczby losowe
    N - długość liczby losowej
    interations - liczba wywołań
    generated_numbers - tablica zawierająca krotki (nr iteracji, wygenerowana liczba)
    iter - licznik iteracji
    '''
    if iterations == 0:
        return generated_numbers
    if N % 2 != 0 and N <= 0:
        print("N musi być parzyste i większe od 0!")
        return
    x = str(seed*seed)
    while(len(x) < 2*N): #wydłużamy liczbę do porządanej wielkości
        x = '0' + x
    z = ''
    i = 0
    mid = len(x)//2
    while(len(z) != N): #skracamy do N środkowych wyrazów
        z = z + x[mid+i]
        z = x[mid-1-i] + z
        i+=1
    z = int(z)
    new_item = [it, z]
    generated_numbers.append(new_item)
    return vonNeumann(z, N, iterations-1, generated_numbers, it+1)
```

```python
N = 4
iterations = 10
seed = int(time.time())
plot = plt.figure(figsize=(20, 20));
plot.suptitle("Przykładowe liczby wygenerowane za pomocą metody von Neumanna dla N=4", fontsize="x-large")
for i in range(3):
    for j in range(3):
        data = np.matrix(vonNeumann(seed, N, iterations, [], 0))
        subplt = plt.subplot(3, 3, i*3+j+1)
        subplt.set_title("Liczba iteracji: " + str(iterations))
        plt.xticks(list(range(0, iterations+1, 5)))
        iterations+=10
        subplt.scatter([data[:,0]], [data[:,1]], marker='o', color='red')
```

## 2. Generatory liniowe
W tym rozdziale zobaczymy kilka przykładów rozkładów generowanych przez zdefiniowane przez nas generatory liniowe. Zachęcam do sprawdzania różnych parametrów, pozostawiam to jako ćwiczenie :). Warto zauważyć, że małe wartości m nie są zbyt ciekawe, ponieważ nie generują ciągów o odpowiednim okresie (przykład poniżej). Podobna sytuacja dotyczy $a=1$, ponieważ zacyklimy się po jednym kroku.

### Ogólna funkcja używana do generacji

```python
def generate_random_number_mul(a, seed, c, m, N):
    '''
    a - parametr "a" generatora
    seed - X_0 dla pierwszego wywołania
    c - parametr "c" generatora
    m - parametr m generatora
    N - liczba liczb do wygenerowania
    '''
    generated_numbers = list()
    i = 0
    next_item = (a*seed + c) % m
    next_item = [i, next_item]
    generated_numbers.append(next_item)
    i+=1
    while i != N:
        next_item = np.int((a*generated_numbers[i-1][1] + c) % m)
        next_item = [i, next_item]
        generated_numbers.append(next_item)
        i+=1
    return generated_numbers
```

### C. Generator oparty na rejestrach przesuwnych (implementacja z ksiązki [2] str. 38)
Korzysta on z trzech generatorów o okresach $(2^{28} -1), (2^{29}-1)$ i $(2^{31}-1)$. Jego okres to iloczyn okresów poszczególnych generatorów składowych. Jak widać na drugim wykresie, generatory tego typu nie posiada struktury przestrzennej, tak jak generatory multiplikatywne.

```python
def rshift(val, n): return val>>n if val >= 0 else (val+0x100000000)>>n
s1 = np.uint32(100)
s2 = np.uint32(150)
s3 = np.uint32(200)
def initial():
    '''
    s1, s2, s3 - liczby startowa
    '''
    global s1, s2, s3
    B = np.uint32(s1) << np.uint32(9)
    B = np.uint32(B) ^ np.uint32(s1)
    B = np.uint32(B) << np.uint32(4)
    s1 = (np.uint32(s1) <<  4) ^ rshift(np.uint32(B), 28)
    B = np.uint32(s2) << np.uint32(2)
    B = np.uint32(B) ^ np.uint32(s2)
    B = np.uint32(B) << np.uint32(3)
    s2 = (np.uint32(s2) <<  3) ^ rshift(np.uint32(B), 29)
    B = np.uint32(s3) << np.uint32(6)
    B = np.uint32(B) ^ np.uint32(s3)
    B = np.uint32(B) << np.uint32(1)
    s3 = (np.uint32(s3) <<  1) ^ rshift(np.uint32(B), 31)
def tausworth_gen():
    global s1, s2, s3
    b = np.uint32(np.uint32(s1 << 9) ^ s1) << 4
    s1 = np.uint32(s1 << 13) ^ np.uint32(rshift(b, 19))
    b = np.uint32(np.uint32(s2 << 2) ^ s2) << 3
    s2 = np.uint32(s2 << 20) ^ np.uint32(rshift(b, 12))
    b = np.uint32(np.uint32(s3 << 6) ^ s3) << 1
    s3 = np.uint32(s3 << 17) ^ np.uint32(rshift(b, 15))
    return np.double((s1^s2^s3)*2.3283064365e-10)
```

```python
initial()
plot = plt.figure(figsize=(15, 15));
plot.suptitle("Generator Tauswortha", fontsize="x-large")
subplt = plt.subplot(1, 2, 1)
data = []
for i in range(0, 10**3):
    data.append(tausworth_gen())
data = np.array(data)
plt.scatter(np.arange(0, 10**3), data)
subplt = plt.subplot(1, 2, 2, aspect='equal')
data = data.reshape(500, 2)
plt.scatter(data[:,0], data[:,1])
plt.show()
```

### D. Generatory Fibonacciego
Jak widać na pierwszym z poniższych wykresów, ciąg reszt mod m z liczb Fibonacciego sam w sobie jest dosyć dobrym generatorem liczb losowych. Wszystko zależy od wyboru liczby m.

```python
def fib(X0, X1, n, m):
    '''
    X0, X1 - startowe wyrazy
    n - liczba liczb do wygenerowania
    m - modulo
    '''
    generated_numbers = [[1, X0], [2, X1]]
    F1 = X0
    F2 = X1
    for i in range(3, n+1):
        F3 = (F1+F2) % m
        F1 = F2
        F2 = F3
        generated_numbers.append([i, F3])
    return generated_numbers
```

```python
plot = plt.figure(figsize=(10, 10));
plot.suptitle("Generator oparty na liczbach Fibonacciego dla m=2^31 - 1", fontsize="x-large")
data = np.array(fib(1, 1, 10**4, 2**31 - 1))
subplt = plt.subplot(1, 2, 1)
plt.scatter(data[:,0], data[:,1]/(2**31 - 1))
subplt = plt.subplot(1, 2, 2)
plt.hist(data[:,1]/(2**31 - 1))
plt.show()
```

## 3. Generatory nieliniowe
Poniżej zamieszczam implementację dwóch omawianych przeze mnie generatorów nieliniowych, oraz kilka wykresów sprawdzające ich własności. Warto zauważyć, że wartości m nie mogą być zbyt duże, ponieważ obliczanie (m-2) potęgi będzie czasochłonne.

### A. Generator Eichenauera-Lehna
Jak widać poniżej, generator dość dobrze przybliża rozkład U(0,1). Można też zauważyć pewną strukutrę dla punktów $(U_i, U_{i+1})$.

```python
def eichenauer_lehn(a, X0, b, m, n):
    '''
    a - mnożnik
    X0 - wyraz początkowy
    b - czynnik dodawany
    m - liczba modulo (musi być pierwsza)
    n - liczba liczb do wygenerowania
    '''
    generated_numbers = []
    for i in tqdm.tqdm(range(0, n)):
        next_x = 0
        if X0 != 0:
            next_x = X0**(m-2)
        X0 = (a*next_x + b) % m
        generated_numbers.append([i, X0])
    return generated_numbers
```

```python
plot = plt.figure(figsize=(15, 15));
plot.suptitle("Generator Eichenauera-Lehna", fontsize="x-large")
data = np.array(eichenauer_lehn(15, int(time.time()), 97, 1031, 10**4))
subplt = plt.subplot(1, 3, 1)
plt.scatter(data[:,0], data[:,1]/1031)
subplt = plt.subplot(1, 3, 2)
plt.hist(data[:,1]/1031)
subplt = plt.subplot(1, 3, 3, aspect='equal')
data = data[:,1].reshape(5000, 2)
plt.scatter(data[:,0]/1031, data[:,1]/1031)
plt.show()
```

### B. Generator Eichenauera-Hermanna

```python
def eichenauer_hermann(a, n0, b, m, n):
    '''
    a - mnożnik
    X0 - wyraz początkowy
    b - czynnik dodawany
    m - liczba modulo (musi być pierwsza)
    n - liczba liczb do wygenerowania
    '''
    generated_numbers = []
    for i in tqdm.tqdm(range(0, n)):
        next_x = 0
        n0 = a*n0+i + b
        if n0 != 0:
            next_x = n0**(m-2)
        n0 = next_x % m
        generated_numbers.append([i, n0])
    return generated_numbers
```

```python
plot = plt.figure(figsize=(15, 15));
plot.suptitle("Generator Eichenauera-Hermanna", fontsize="x-large")
data = np.array(eichenauer_hermann(15, int(time.time()), 0, 4111, 10**3))
subplt = plt.subplot(1, 3, 1)
plt.scatter(data[:,0], data[:,1]/1031)
subplt = plt.subplot(1, 3, 2)
plt.hist(data[:,1]/4111)
subplt = plt.subplot(1, 3, 3, aspect='equal')
data = data[:,1].reshape(500, 2)
plt.scatter(data[:,0]/4111, data[:,1]/4111)
plt.show()
```

## 4. Generatory o dowolnym rozkładzie
W tym rozdziale przedstawię kilka metod przekształcania rozkładu U(0, 1) na inne. W implementacjach użyjemy generatora z biblioteki random. Więcej przykładów można znaleźć w [2] rozdział 3.2.

### A. Rozkład wykładniczy
Najlepszą metodą generowania rozkładu wykładniczego jest metoda odwracania dystrybuatny, dlatego się na niej skupimy. Rozkład wykładniczy $E(\theta, \lambda)$ ma gęstość $f(x)=\frac{1}{\lambda}\exp(-\frac{x-\theta)}{\lambda})$. Jeśli zmienna losowa $Y~E(\theta, \lambda)$ to zmienna $X=(Y-\theta)/\lambda)$ ma rozkład $E(0,1)$, więc skupimy się na generowaniu liczb o takim rozkładzie. Dystrybuanta zmiennej X wyraża się wzorem $F(x)=1-e^{-x}$. Teraz jeśli U ma rozkład jednostajny U(0, 1) to $F^{-1}(x)=-\ln(U)$.

```python
def gen_exp(n, inline=False):
    '''
    n - liczba liczb do wygenerowania
    '''
    generated_numbers = []
    pbar = 0
    if not inline:
        pbar = tqdm.tqdm(total=n)
    for i in range(0, n):
        u = np.random.rand()
        x = -np.log(u)
        generated_numbers.append([i, x])
        if not inline:
            pbar.update(1)
    if not inline:
        pbar.close()
    return generated_numbers
theta = 0
lambd = 1
data = np.array(gen_exp(10**5))[:,1]
count, bins, ignored = plt.hist(data, 100, density=True)
plt.plot(bins, (1/lambd)*np.exp(-((bins-theta)/lambd)), 
         linewidth=2, color='r')
plt.show()
```

### B. Rozkład normalny
Rozkład normalny o parametrach $N(\mu, \sigma)$ można uzyskać ze zmiennej losowej $X\~N(0, 1)$ za pomocą przekształcenia $Y=(X\cdot \sigma) + \mu$.

#### Metoda odwracania dystrybuanty
W notatce wspomniane było, że czasami odwracanie dystrybuanty jest trudne w sensie numerycznym. Tak właśnie jest w tym przypadku, dlatego skorzystamy z przybliżenia funkcji odwrotnej do dystrybuanty N(0,1) podanej przez Odeha I Evansa w 1974 roku.

```python
def gen_norm_cdf(n, inline=False):
    '''
    n - liczba liczb do wygenerowania
    '''
    def horner(x, coefficients):
        result = 0
        for coefficient in coefficients:
            result = result * x + coefficient
        return result
    M_coeffs = [0.0038560700634, 0.103537752850, 0.531103462366, 0.588581570495, 0.0993484626060]
    L_coeffs = [0.0000453642210148, 0.0204231210245, 0.342242088547, 1, 0.322232431088]
    generated_numbers = []
    i = 0
    if not inline:
        pbar = tqdm.tqdm(total=n)
    while i < n:
        u = np.random.rand()
        t = np.sqrt(-2*np.log(u))
        if u > 10**(-20) and u < 0.5:
            g = t - (horner(t, L_coeffs)/horner(t, M_coeffs))
            u = np.random.rand()
            if u > 0.5:
                g=-g
            generated_numbers.append([i, g])
            i+=1
            if not inline:
                pbar.update(1)
        elif u <= 0.5 and u < 1 - 10**(-20):
            g = t - (horner(t, L_coeffs)/horner(t, M_coeffs))
            g = (-1*g)*(1-u)
            u = np.random.rand()
            if u > 0.5:
                g=-g
            generated_numbers.append([i, g])
            i+=1
            if not inline:
                pbar.update(1)
    if not inline:
        pbar.close()
    return generated_numbers
```

```python
sigma = 1
mu = 0
data = np.array(gen_norm_cdf(10**5))[:,1]
count, bins, ignored = plt.hist(data, 100, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), 
         linewidth=2, color='r')
plt.show()
```

#### Metoda eliminacji
Będziemy generować X z dodatniej połowy rozkładu normalnego (gęstość takiego rozkładu to $f(x)=\sqrt{2/\pi}\exp(-x^2/2)$), a następnie z prawdopodobieństwem $p=\frac{1}{2}$ mnożyć razy $-1$. Skorzystamy z rozkładu wykładniczego, który jest łatwo wygenerować, co widzieliśmy wyżej. Niech $g(x)=e^{-x}$ będzie gęstością dominującą. Wtedy nasza stała c to $c=\sqrt{2e/\pi}$.

```python
def gen_norm_elim(n, inline=False):
    '''
    n - liczba liczb do wygenerowania
    '''
    generated_numbers = []
    i = 0
    if not inline:
        pbar = tqdm.tqdm(total=n)
    while i < n:
        while True:
            x = gen_exp(1, inline=True)
            x = x[0][1]
            u = np.random.rand()
            if np.sqrt(2*np.e/np.pi)*u*np.exp(-x) <= np.sqrt(2/np.pi)*np.exp((-1*x*x)/2):
                break
        u = np.random.rand()
        if u >= 0.5:
            x=-x
        generated_numbers.append([i, x])
        i+=1
        if not inline:
            pbar.update(1)
    if not inline:
        pbar.close()
    return generated_numbers
```

```python
sigma = 1
mu = 0
data = np.array(gen_norm_elim(10**5))[:,1]
count, bins, ignored = plt.hist(data, 100, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), 
         linewidth=2, color='r')
plt.show()
```

## 5. Testowanie zgodności generatorów.
W tym rozdziale testować będziemy nasze generatory za pomocą metod omówionych w notatce. Dla testu Kołomogorowa skorzystamy z funkcji bibliotecznych pakietu scipy. Jeżeli testujemy generator dla danego poziomu istotoności $\alpha$, to powinien on nie przechodzić testu średnio raz na $\frac{1}{\alpha}$ przypadków. Taki odsetek niepowodzeń, nie powoduje odrzucenia generatora.

### A. Test chi-kwadrat

```python
def chi_square(data, k=10, sig=0.05):
    '''
    Funkcja oblicza wartość testu chi-kwadrat dla rozkładu jednostajnego na przedziale [0, 1)
    data - wygenerowane liczby
    k - liczba przedziałów
    sig - poziom istotoności testu
    '''
    intervals = np.arange(0, 1.1, 1/k)
    inter = []
    for i in range(k):
        inter.append([intervals[i], intervals[i+1]])
    all_obs = []
    for j in range(0, k):
        x = data[data > inter[j][0]]
        x = x[x <= inter[j][1]]
        all_obs.append(x.shape[0])
    all_obs=np.array(all_obs).reshape(k, 1)
    all_exp = np.zeros([k, 1])
    all_exp[all_exp == 0] = data.shape[0]/k
    chi2_test = (((all_obs-all_exp)**2)/all_exp).sum()
    crit = stats.chi2.ppf(q=1-sig, df=k-1)
    if(chi2_test <= crit):
        return 1
    else:
        return 0
```

```python
#testowanie n=10 prób
for i in range(0, 10):
    data = np.array(generate_random_number_mul(234237, int(time.time()), 0, 105, 10**5))[:,1]/105
    if chi_square(data):
        print("Spełniono test")
    else:
        print("Nie spełniono testu")
```

```python
#Testowanie generatora liniowego a=16807, c=0, m=2^32, który powinien przejść testy
zeros = 0
for i in tqdm.tqdm(range(0, 10**3)):
    data = np.array(generate_random_number_mul(16807, int(time.time()), 0, 2**32, 10**5))[:,1]/2**32
    if chi_square(data) == 0:
        zeros+=1.
all_obs=[zeros, 10**3-zeros]
all_obs=np.array(all_obs).reshape(2, 1)
all_exp = np.zeros([2, 1])
all_exp[0] = 0.05*(10**3)
all_exp[1] = 10**3-all_exp[0]
chi2_test = (((all_obs-all_exp)**2)/all_exp).sum()
crit = stats.chi2.ppf(q=1-0.05, df=1)
if(chi2_test <= crit):
    print(1)
else:
    print(0)
```

### C. Test pokerowy
Zdefinujemy najpierw funkcje prawdopodobieństw odpowiednich piątek, później użyjemy ich do ewaluacji ich rozkładu przy pomocy testu chi-kwadrat

```python
def bust(k):
    if k>=5:
        return ((k-1)*(k-2)*(k-3)*(k-4))/k**4
    else:
        return 0
def pair(k):
    if k>=4:
        return (10*(k-1)*(k-2)*(k-3))/k**4
    else:
        return 0
def double_pair(k):
    if k>=3:
        return (15*(k-1)*(k-2))/k**4
    else:
        return 0
def triple(k):
    if k>=3:
        return (10*(k-1)*(k-2))/k**4
    else:
        return 0
def full(k):
    if k>=3:
        return (10*(k-1))/k**4
    else:
        return 0
def quad(k):
    if k>=2:
        return (5*(k-1))/k**4
    else:
        return 0
def five(k):
    if k>=2:
        return 1/k**4
    else:
        return 0
def identify_tuple(t, k):
    how_many = []
    k = np.arange(0, k)
    for i in k:
        how_many.append(t[t == i].shape[0])
    how_many = np.array(how_many)
    if how_many[how_many == 5].shape[0] == 1:
        return 6
    elif how_many[how_many == 4].shape[0] == 1:
        return 5
    elif how_many[how_many == 3].shape[0] == 1 and how_many[how_many == 2].shape[0] == 1:
        return 4
    elif how_many[how_many == 3].shape[0] == 1:
        return 3
    elif how_many[how_many == 2].shape[0] == 2:
        return 2
    elif how_many[how_many == 2].shape[0] == 1:
        return 1
    elif how_many[how_many == 1].shape[0] == 5:
        return 0
def poker_test(data, k, a, b, sig=0.05):
    '''
    data - próbka wielkości n t.że n % 5 = 0
    k - liczba przedziałów
    [a, b) - przedział wartości zmiennej losowej
    '''
    intervals = np.arange(a, b, 1/k)
    inter = []
    Y = []
    for i in range(k):
        inter.append([intervals[i], intervals[i+1]])
    for i in range(data.shape[0]): #przygotowujemy ciąg Y
        for j in range(k):
            if data[i][0] > inter[j][0] and data[i][0] <= inter[j][1]:
                Y.append(j)
                break
    Y = np.array(Y)
    Y = Y.reshape(data.shape[0]//5, 5)
    Y = np.sort(Y)
    probs = np.array([bust(k), pair(k), double_pair(k), triple(k), full(k), quad(k), five(k)])
    probs*=Y.shape[0]
    all_exp=probs.reshape(7, 1)
    all_obs = np.zeros(7, dtype=np.int)
    for i in Y:
        type_of_tuple = identify_tuple(i, k)
        all_obs[type_of_tuple]+=1
    all_obs = all_obs.reshape(7, 1)
    chi2_test = (((all_obs-all_exp)**2)/all_exp).sum()
    crit = stats.chi2.ppf(q=1-sig, df=k-1)
    if(chi2_test <= crit):
        return 1
    else:
        return 0
```

```python
poker_test(np.random.rand(10000, 1), 10, 0, 1.1)
```
