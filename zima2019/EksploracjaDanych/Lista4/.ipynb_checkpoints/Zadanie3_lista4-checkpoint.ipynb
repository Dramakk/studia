```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.cluster import KMeans, MiniBatchKMeans, Birch, DBSCAN
from sklearn.metrics import davies_bouldin_score
import tqdm
import spectral
```

```python
img = spectral.open_image(
    '/gdrive/My Drive/APEX_OSD_Package_1.0/APEX_OSD_V1_calibr_cube.hdr'
)
data = img.load()
A = data.reshape(-1, data.shape[-1])
```

```python
A_std = (A - A.mean()) / A.std()
A_norm = (A - np.min(A)) / (np.max(A) - np.min(A))
```

```python
km = MiniBatchKMeans(n_clusters=6, init="k-means++", n_init=10).fit(A)
X = km.labels_
print(f'Davies bouldin index: {davies_bouldin_score(A, km.labels_)}')
plt.figure(figsize=(15, 10))
plt.imshow(X.reshape(1500,1000), cmap='gist_earth')
```

```python
km = MiniBatchKMeans(n_clusters=9, init="k-means++", n_init=10).fit(A)
X = km.labels_
print(f'Davies bouldin index: {davies_bouldin_score(A, km.labels_)}')
plt.figure(figsize=(15, 10))
plt.imshow(X.reshape(1500,1000), cmap='gist_earth')
```

```python
km = MiniBatchKMeans(n_clusters=12, init="k-means++", n_init=10).fit(A)
X = km.labels_
print(f'Davies bouldin index: {davies_bouldin_score(A, km.labels_)}')
plt.figure(figsize=(15, 10))
plt.imshow(X.reshape(1500,1000), cmap='gist_earth')
```

```python
km = KMeans(n_clusters=12, init="k-means++", n_init=10).fit(A)
X = km.labels_
print(f'Davies bouldin index: {davies_bouldin_score(A, km.labels_)}')
plt.figure(figsize=(15, 10))
plt.imshow(X.reshape(1500,1000), cmap='gist_earth')
```

```python
dbs = DBSCAN(eps=0.3, min_samples=10).fit(A)
X = dbs.labels_
plt.figure(figsize=(15, 10))
plt.imshow(X.reshape(1500,1000), cmap='gist_earth')
```
