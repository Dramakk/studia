```python
import numpy as np
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
w = np.random.rand(100)
x_len = np.sqrt(np.dot(x,x))
x_len
```

```python
x_avg = np.sum(x*w)/np.sum(w)
x_avg
```

```python
dot_prod = np.dot(x, y)
dot_prod
```

```python
temp = x-y
distance = np.sqrt(np.dot(temp, temp))
distance
```

```python
X = np.random.rand(100, 1000)
X_len = np.sqrt(np.sum(X ** 2, axis=0))
X_len
```

```python
X_avg = np.sum(X.T*w, axis=1)/np.sum(w)
X_avg
```

```python
X_dist = np.sqrt(np.sum(((X.T-y) ** 2), axis=1))
X_dist
```

```python
X_dot = np.dot(X.T, y)
X_dot
```
