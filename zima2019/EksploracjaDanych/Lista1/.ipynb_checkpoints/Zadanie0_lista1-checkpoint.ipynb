```python
c = np.arange(-1.0, 1.01, step=0.01, dtype=np.float64) * np.pi
c[100] = 0 #Było coś bardzo bliskie zeru
c
```

```python
e = np.sin(np.arange(1, 101))
e[e <= 0] = 0
e
```

```python
A = np.arange(1, 101)
A = A.reshape(10, 10)
A
```

```python
B = np.zeros((100, 100))
x = np.arange(0, 100)
B[x, x] = np.arange(1, 101)
B[x[1:], x[:-1]] = np.arange(99, 0, step=-1)
B[x[:-1], x[1:]] = np.arange(99, 0, step=-1)
B
```

```python
C = np.triu(np.ones((10, 10)))
C
```

```python
D = np.reshape([np.arange(1, 101), np.arange(1,101)], (2, 100))
D[0] = np.cumsum(D[0])
D[1] = np.cumproduct(D[1])
D
```

```python
E = np.zeros((100, 100))
x = np.ndarray((100, 100))
x[np.arange(0, 100)] = np.arange(1, 101) #Liczmy jak ludzie to znaczy, że macierz zaczyna się na (1, 1)
E[x % x.T == 0] = 1
E
```
