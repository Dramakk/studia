```python
import matplotlib.pyplot as plt
#b
iris_d = np.array(iris.data)
plt.plot(iris.data[:,0], iris.data[:,1], 'ro')
plt.ylabel('sepal width (cm)')
plt.xlabel('sepal length (cm)')
plt.title('Iris')
plt.show()
```

```python
#c
plt.plot(iris.data[:,0], iris.data[:,1], 'ro')
plt.ylabel('sepal width (cm)')
plt.xlabel('sepal length (cm)')
plt.title('Iris')
plt.xlim(3, 9)
plt.ylim(1, 5)
plt.show()
```

```python
#d
plt.plot(iris.data[:,0], iris.data[:,1], 'ro')
plt.ylabel('sepal width (cm)')
plt.xlabel('sepal length (cm)')
plt.title('Iris')
plt.xlim(3, 9)
plt.ylim(1, 5)
plt.xticks(range(3, 10))
plt.yticks(range(1, 6))
plt.show()
```
