```python
from sklearn import datasets
iris = datasets.load_iris()
import matplotlib.pyplot as plt
import numpy as np
```

```python
A = iris.data[iris.target != 1]
plt.ylabel('sepal width (cm)')
plt.xlabel('sepal length (cm)')
plt.title('Iris')
plt.xlim(3, 9)
plt.ylim(1, 5)
plt.xticks(range(3, 10))
plt.yticks(range(1, 6))
plt.scatter(A[:,0], A[:,1],c=iris.target[iris.target != 1])
plt.show()
```

```python
#b
X = np.arange(3, 10)
Y = 2*X-8
plt.plot(X, Y)
plt.ylabel('sepal width (cm)')
plt.xlabel('sepal length (cm)')
plt.title('Iris')
plt.xlim(3, 9)
plt.ylim(1, 5)
plt.xticks(range(3, 10))
plt.yticks(range(1, 6))
plt.scatter(A[:,0], A[:,1], c=iris.target[iris.target != 1])
plt.show()
```

```python
#c
def f(x):
    return (2*x) - 8

setosa = iris.data[iris.target == 0]
virginica = iris.data[iris.target == 2]

setosa_above = setosa[setosa[:,1] >= f(setosa[:,0])]
setosa_below = setosa[setosa[:,1]<f(setosa[:,0])]
virginica_above = virginica[virginica[:,1] >= f(virginica[:,0])]
virginica_below = virginica[virginica[:,1] < f(virginica[:,0])]

X = np.arange(3, 10)
plt.plot(X, f(X))
plt.ylabel('sepal width (cm)')
plt.xlabel('sepal length (cm)')
plt.title('Iris')
plt.xlim(3, 9)
plt.ylim(1, 5)
plt.xticks(range(3, 10))
plt.yticks(range(1, 6))
plt.scatter(setosa_above[:,0], setosa_above[:,1], marker='o', c='green')
plt.scatter(setosa_below[:,0], setosa_below[:,1], marker='o', c='red')
plt.scatter(virginica_above[:,0], virginica_above[:,1], marker='o', c='red')
plt.scatter(virginica_below[:,0], virginica_below[:,1], marker='o', c='green')
plt.savefig("zadanie2a.png")
plt.show()
```

```python
#d
def f(x):
    return (1.25*x) - 3.5

setosa = iris.data[iris.target == 0]
virginica = iris.data[iris.target == 2]

setosa_above = setosa[setosa[:,1] >= f(setosa[:,0])]
setosa_below = setosa[setosa[:,1]<f(setosa[:,0])]
virginica_above = virginica[virginica[:,1] >= f(virginica[:,0])]
virginica_below = virginica[virginica[:,1] < f(virginica[:,0])]

X = np.arange(3, 10)
plt.plot(X, f(X))
plt.ylabel('sepal width (cm)')
plt.xlabel('sepal length (cm)')
plt.title('Iris')
plt.xlim(3, 9)
plt.ylim(1, 5)
plt.xticks(range(3, 10))
plt.yticks(range(1, 6))
plt.scatter(setosa_above[:,0], setosa_above[:,1], marker='o', c='green')
plt.scatter(setosa_below[:,0], setosa_below[:,1], marker='o', c='red')
plt.scatter(virginica_above[:,0], virginica_above[:,1], marker='o', c='red')
plt.scatter(virginica_below[:,0], virginica_below[:,1], marker='o', c='green')
plt.savefig("zadanie2b.png")
plt.show()
```
