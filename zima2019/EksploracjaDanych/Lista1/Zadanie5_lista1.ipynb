```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(1,11)
Y = np.array([10, 10, 11, 12, 18, 18, 19, 26, 19, 26])
```

```python
plt.ylabel('Y')
plt.xlabel('X')
plt.plot(X, Y)
plt.show()
```

```python
def f(a, b, x):
    return (a*x) + b
f_Y = f(2, 5, X)

plt.ylabel('Y')
plt.xlabel('X')
plt.plot(X, Y, '.')
plt.plot(X, f_Y)
plt.savefig("zadanie3a.png")
plt.show()
```

```python
def dist_from_f(a, b, x, y):
    return np.abs((-a*x)+y-b)/np.sqrt((-a) ** 2 + 1)
points = np.column_stack([X, Y])
points_dist_from_f = dist_from_f(2, 5, points[:,0], points[:,1])
sum_of_dists = np.sum(points_dist_from_f)
sum_of_dists
```

```python
f_Y = f(2, 6, X)

plt.ylabel('Y')
plt.xlabel('X')
plt.plot(X, Y)
plt.plot(X, f_Y)
plt.savefig("zadanie3b.png")
plt.show()

points = np.column_stack([X, Y])
points_dist_from_f = dist_from_f(2, 6, points[:,0], points[:,1])
sum_of_dists = np.sum(points_dist_from_f)
sum_of_dists
```
